```python
import numpy as np  # np mean, np random
import pandas as pd  # read csv, df manipulation
import re
import plotly.express as px  # interactive charts
import streamlit as st  # data web app development
```

```python
#Importar CSV
df=pd.read_csv("PIDatabase.csv")
df.head
```

```python
#Eliminar columnas inecesarias
PIdf = df.drop(['Fecha', 'Actividad Actual', 'CampusAdministrador', 'OportunidadesAprobadas', 'Núm. MatAprobadas', 'Núm. MatTotales', 'PeriodoAcadémico','Nombre','Ap Paterno','Ap Materno','Instancia','Periodo Solicitud','Promedio','PuntajeAsignación'], axis=1)
PIdf
```

```python
#Dejar solo "Asignad@"
variables = ['Asignado','Asignada'] 
estatus = '|'.join(variables)
asignado = PIdf[PIdf['Estatus'].str.match(estatus)]
PIdf=pd.DataFrame(asignado)
PIdf
```

```python
#Valores Nulos en "Programa"
PIdf.fillna({'Programa': 'LIN11'}, inplace=True)
PIdf.fillna({'Programa': {0: 'LIN11'}})
```

```python
#Eliminar duplicados de matrículas
PIdf_2= PIdf.drop_duplicates(subset=['Matrícula'])
PIdf_2
#26465
```

```python
#Creación columna de País
db1=PIdf_2["OportunidadAsignada"].str.split("-", expand=True)
Pais=db1[0]
PIdf_2["Pais"]=Pais
PIdf_2
```

```python
#Creación columna de Tipo de Intercambio
db1=PIdf_2["OportunidadAsignada"].str.split("-", expand=True)
Clave=db1[1]
PIdf_2["Tipo"]=Clave
PIdf_2
```

```python
#Importar CSV "ProgramasTec"
df2=pd.read_csv("ProgramasTec.csv")
df2.head
```

```python
#Merge de CSV´s
PIdf_2=pd.merge(left=PIdf_2, right=df2, left_on=["Programa"], right_on=["Clave"])
PIdf_2.columns
PIdf_2
```
